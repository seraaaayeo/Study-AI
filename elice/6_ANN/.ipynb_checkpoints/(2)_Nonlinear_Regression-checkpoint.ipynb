```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from mpl_toolkits.mplot3d import Axes3D
```

```python
print(tf.__version__)
```

```python
np.random.seed(100)
tf.random.set_seed(100)
```

```python
# 데이터 생성
x_data = np.linspace(0, 10, 100)
y_data = 1.5 * x_data**2 -12 * x_data + np.random.randn(*x_data.shape)*2 + 0.5
```

```python
# 1. 신경망 모델 생성
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(20, input_dim=1, activation='relu'), #Input layer: input_dim: 입력되는 데이터의 차원.
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(1) #Output layer: 출력해야 하는 결과값의 개수만큼 노드의 개수를 설정.
])

# 2. 모델 학습 방법 설정
model.compile(loss='mean_squared_error', optimizer='adam')

# 3. 모델 학습 
model.fit(x_data, y_data, epochs=500, verbose=2) #verbose: 모델 학습 과정 정보를 얼마나 자세하게 출력할 것인가(0, 1, 2)

# 4. 학습된 모델을 사용하여 예측값 생성 및 저장
predictions = model.predict(x_data)

# 데이터 출력
plt.scatter(x_data,y_data)
plt.savefig('data.png')

# 곡선형 분포 데이터와 예측값 출력
plt.scatter(x_data,predictions, color='red')
plt.savefig('prediction.png')
```
