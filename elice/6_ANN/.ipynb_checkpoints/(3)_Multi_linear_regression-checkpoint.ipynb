````python
from __future__ import absolute_import, division, print_function, unicode_literals

import tensorflow as tf
import numpy as np
import random

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
````

````python
import csv
import numpy as np

def read_data():
    csvreader = csv.reader(open("/content/drive/My Drive/Colab Notebooks/data/Advertising.csv"))

    x = []
    y = []

    next(csvreader)
    for line in csvreader :
        x_i = [ float(line[1]), float(line[2]), float(line[3]) ]
        y_i = float(line[4])
        x.append(x_i)
        y.append(y_i)

    X = np.array(x)
    Y = np.array(y)
    
    return X,Y
````

````python
tf.random.set_seed(123)
np.random.seed(123)
````

````python
X, Y = read_data()
print(X.shape)
print(Y.shape)
````

### 학습용 데이터와 테스트용 데이터를 8:2로 분리

````python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=0)
````

## MSE가 1 이하인 모델 구현

### 모델 구현

````python
model = tf.keras.models.Sequential([
    # Input Layer
    tf.keras.layers.Dense(20, input_dim = 3, activation='relu'),
    # MSE 값을 1 이하로 낮출 수 있도록 여러 층의 hidden layer를 추가해보세요.
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(20, activation='relu'),
    # Output Layer
    tf.keras.layers.Dense(1)
])
````

### 모델 학습 방법 설정

````python
model.compile(loss= 'mean_squared_error', optimizer='adam')
````

### epoch 설정하고 모델 학습

````python
model.fit(x_train, y_train, epochs=800)
````

### 학습된 신경망 모델을 사용하여 예측값 생성 및 loss 출력

````python
predicted = model.predict(x_test)
mse_test = mean_squared_error(predicted, y_test)
print("MSE on test data: {}".format(mse_test))
````

#### 느낀 점
* 모델 구성할 때 활성함수를 sigmoid와 relu를 섞으면 MSE가 확 높아졌다.
* 모델 layer가 깊어진다고 능사는 아니더라. layer가 5층일 땐 오히려 MSE가 높아졌다.overfitting 때문인 듯.
* 마찬가지로 epoch가 많다고 능사 또한 아니다. layer 5층일 때 epoch 500일 때 보다 epoch 1000일 때 오히려 MSE가 높았다.
* 결국 layer의 깊이, 활성함수, epoch 전부 적절하게 조합하는 것이 모델 성능의 관건.

* layer 4층, 활성함수 relu, 에포치 800일 때
    ```
    MSE on test data: 0.9628272407763125
    ```
* layer 4층, 활성함수 relu, 에포치 1000일 때
    ```
    MSE on test data: 0.9170071607297684
    ```
* layer 5층, 활성함수 relu, 에포치 1000일 때
    ```
    MSE on test data: 1.7549667796917454
    ```
