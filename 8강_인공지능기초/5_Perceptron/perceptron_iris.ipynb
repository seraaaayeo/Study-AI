```python
import sklearn
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
```

```python
def load_irisdata():
    iris = datasets.load_iris()
    
    X = iris.data #붓꽃 데이터의 4가지 feature
    Y = iris.target #붓꽃 데이터의 3가지 품종 class
    
    return X, Y
```

```python
def main():
    X, Y = load_irisdata()
    print(X)
    print(Y)
    
    #X, Y 데이터를 훈련용 데이터와 테스트 데이터로 분류(8:2)
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
    
    #sklearn의 Perceptron 클래스를 사용하여 train 데이터 학습
    #max_iter=모델 학습 최대 횟수.얼마나 많은 데이터로 학습을 할 것인가 
    #eta0 = 모델 학습 속도.한 번 iteration 할 때 마다 얼마나 빨리 학습을 하는가
    perceptron = Perceptron(max_iter=120, eta0=1) 
    perceptron.fit(x_train, y_train)
    
    #test 데이터에 대한 예측값 생성
    pred = perceptron.predict(x_test)
    print("Test 데이터에 대한 정확도: %f" %accuracy_score(pred, y_test))
    
    return x_train, x_test, y_train, y_test, pred
```

```python
if __name__ == "__main__":
    main()
```
